```python
loc1 = np.array([50, 60, 70, 70])
loc2 = np.array([40, 50, 60, 60])
loc3 = np.array([40, 50, 50, 50])
loc4 = np.array([30, 40, 40, 50])
loc5 = np.array([30, 40, 40, 40])
loc6 = np.array([30, 30, 40, 50])


av = (loc1+loc2+loc3+loc4+loc5+loc6)/6
```

```python
fig = plt.figure(figsize=(6,5))

plt.plot(x, loc1)
plt.plot(x, loc2)
plt.plot(x, loc3)
plt.plot(x, loc4)
plt.plot(x, loc5)
plt.plot(x, loc6)
plt.xticks([0,1,2,3], ["E14.5", "E15.5", "E16.5", "E17.5"],fontsize=18)
plt.yticks([20, 30, 40, 50, 60, 70])
plt.ylim([5,75])
plt.xlim([-0.25, 3.25])
plt.legend(['loc 1', 'loc 2', 'loc 3', 'loc 4', 'loc 5', 'loc 6'],loc=8, ncol=3, columnspacing=0.7, handlelength=1)
plt.ylabel('thickness $[\mu m]$', fontsize=20)

plt.savefig("thickness.png", bbox_inches='tight', dpi=600)
```
